````python
# TCP      (MyTCPServer.py)

class TCPServer:
    # TCP    
    def __init__(self, port):
        import socket
        self.sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        self.sock.bind(('', port)) #  
        self.sock.listen(1) # 
        self.connected = False
        
    #  
    def disconnect(self):
        self.connected = False
        self.c_sock.close()
    
    #   
    def accept(self): # 
        self.c_sock, self.c_addr = self.sock.accept()
        self.connected = True
        return self.c_sock, self.c_addr #   

    #  
    def send(self, msg):
        if self.connected:
            self.message = msg
            self.c_sock.send(self.message.encode())
            return True #   
        else:
            return False #   
        
    #  
    def receive(self, r_sock=None):
        if not r_sock:
            r_sock = self.c_sock #   
        try:
            data = r_sock.recv(1024)
            if not data:
                self.disconect()
                return None
            return data.decode() #   
        except:
            self.disconnect()
            return None
````

````python
for i in range(10): #10 
    delay = 0.1  #0.1  
    sub_total = 0 # total delay time for one trial
    data = 'Hello message ' + str(i)

    while True:
        print("\nSending:", data)
        c_sock.send(data.encode())
        print(f'Waiting up to {delay} seconds for a reply')
        c_sock.settimeout(delay) # 
        try:
            resp = c_sock.recv(BUFFER) # 
        except timeout: # 
            print("No response!!!")
            delay *= 2  #  2 
            if delay > 2.0: # 
                print('The server seems to be down')
                break 
        else:
            sub_total += delay 
            print(resp.decode(), "received")
            break #
        
    total_delay += sub_total

print(f"Average Delay Time={total_delay/10.:.2f}") # average delay time
````
